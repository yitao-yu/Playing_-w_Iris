```python
import numpy as np
import math
```

```python
from decimal import Decimal as D
from decimal import *
```

```python
#Data Set#
x1 = [D("1.5"), D("1.7")]
x2 = [D("2.0"),D("1.9")]
x3 = [D("1.6"), D("1.8")]
x4 = [D("1.2"), D("1.5")]
x5 = [D("1.5"),D("1.0")]
dataset = [x1,x2,x3,x4,x5]
x = [D("1.4"), D("1.6")]
```

```python
setofdiff = []
for i in dataset:
    a = np.array(i)
    x = np.array(x)
    diff = a - x
    setofdiff.append(diff)
```

```python
setofdiff
```

```python
def Euclidean(a):
    d = []
    for i in a:
        n = D("0")
        for j in i:
            n += j**2
        n = D(math.sqrt(n))
        d.append(n)
    return d
```

```python
def Manhattan(a):
    d = []
    for i in a:
        n = D("0")
        for j in i:
            n += j
        d.append(n)
    return d
```

```python
def supremum(a):
    d = []
    for i in a:
        n = D("0")
        for j in i:
            if(abs(j)>abs(n)):
                n = abs(j)
        d.append(n)
    return d
```

```python
def length(v):
    n = D("0")
    for i in v:
        n += i**2
    n = n.sqrt()
    return n
```

```python
def cosine(data,x):
    d = []
    for i in data:
        upper = D(np.dot(i,x))
        lower = D(length(i) * length(x))
        d.append(upper/lower)
    return d
```
