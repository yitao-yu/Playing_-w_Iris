```python
#Trimming the data frame and some information stored#
#And also cleasning the species by deleting "Iris-"#
#In case that we might use the original dataframe later, we will operate on second alternative dataframe df#
```

```python
for i,row in df.iterrows():
    ar = df["Species"][i].split("-")
    df["Species"][i] = ar[1]
```

```python
del df["Id"]
```

```python
#Let's plot the dataset!#
#What I'm trying to do here: 
#Plot a scatter plot matrix with Iris data
# points in scatter plot colored in terms of the Species#

#Reference1: The idea of plotting data into scatter plot comes from professor lecture and wikipedia picture describing Iris data#
```

```python
colors = []
arr = ['setosa', 'versicolor', 'virginica']
for i,row in df.iterrows():
    s = df["Species"][i]
    if s == arr[0]:
        colors.append("red")
    elif s == arr[1]:
        colors.append("blue")
    elif s == arr[2]:
        colors.append("green")
    else:
        print(s)
        print("ERROR: NOOOOT A SPECIES OF IRIS")
        
colors
```

```python
plot = pd.plotting.scatter_matrix(df,alpha = 1,figsize = (11,11),color=colors,diagonal = 'kde')
print("red for setosa")
print("blue for versicolor")
print("green for virginica")
#Using pandas honestly feels like cheating#
```
